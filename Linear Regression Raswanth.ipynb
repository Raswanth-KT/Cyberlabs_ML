```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
#Complete Dataset of Training and Testing
data1 = pd.read_csv("Linear_train.csv")
data2 = pd.read_csv("Linear_test.csv")
```

```python
M = data1.drop(data1.columns[[0, 21]], axis=1)  #Features of Training set
N = data2.drop(["Unnamed: 0", "label"], axis=1)  #Features of Testing set
```

```python
train_data = data1.values
K = train_data[:, -1].reshape(train_data.shape[0], 1)  #Labels of Training dataset
```

```python
test_data = data2.values
K_test = test_data[:, -1].reshape(test_data.shape[0], 1)   #Labels of Testing dataset 
```

```python
mean_train = np.mean(M, axis=0)
std = np.std(M, axis=0)
train_f = (M - mean_train)/std
train_f

mean_test = np.mean(N, axis=0)
std = np.std(N, axis=0)
test_f = (N - mean_test)/std
test_f
```

```python
 #Adding column of ones before all the Features
train_f = np.vstack((np.ones((train_f.shape[0], )), train_f.T)).T
test_f = np.vstack((np.ones((test_f.shape[0], )), test_f.T)).T
```

```python
def LinearRegression(M, K, lr, iter):
    m = K.size
    theta = np.zeros((M.shape[1], 1))
    cost_list = []
    for i in range(iter):        
        y_p = np.dot(M, theta)
        
        #Cost Function
        cost = (1/(2*m))*np.sum(np.square(y_p - K))
        
        #Gradient Descent
        d_theta = (np.dot(M.T, y_p - K))*(1/m)
        theta = theta - lr*d_theta
        
        cost_list.append(cost)
        if (i%(iter/10) == 0):
            print("Cost is :", cost)
        
    return theta, cost_list
```

```python
iter = 100
lr = 0.05
theta, cost_list = LinearRegression(test_f, K_test, lr = lr, iter = iter)
```

```python
arg = np.arange(0, iter)
plt.plot(arg, cost_list)
plt.show
```

```python
y_p = np.dot(test_f, theta)
error = (1/test_f.shape[0])*(np.sum(np.abs(y_p - K_test)))
error
```

```python
m = K_test.size
rmse = np.sqrt(np.sum(np.square(y_p - K_test))/m)
rmse
```
